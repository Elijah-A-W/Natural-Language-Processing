**ONE HOT REPRESENTATION**

```
# Xtrain
```

```
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
#DATA CLEANINING


from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```


corpus = []

for i in range(0, len(Xtrain)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', Xtrain['safe_text'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```
corpus1 = []

for i in range(0, len(Xtest)):
    print(i)
    review1 = re.sub('[^a-zA-Z]', ' ', Xtest['safe_text'][i])
    review1 = review1.lower()
    review1 = review1.split()

    review1 = [ps.stem(word) for word in review1 if not word in stopwords.words('english')]
    review1 = ' '.join(review1)
    corpus1.append(review1)



```

```
corpus
```

```
# One hot encoding 
Onehot = [one_hot(words,Vocabulary_size)for words in corpus]
Onehot



```

**Embedding Representation**

```
sent_length = 20
embedded_doc = pad_sequences(Onehot, padding='pre', maxlen= sent_length)
print(embedded_doc)
```

```
len(embedded_doc)
```

**CREATING A MODEL**

```
embedding_vectors_features = 40
model = Sequential()
model.add(Embedding(Vocabulary_size,embedding_vectors_features, input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss='binary_crossentropy',optimizer = 'adam', metrics=['accuracy'])
print(model.summary())
```

```
len(embedded_doc),Ytrain.shape
```

```
import numpy as np

x_fin = np.array(embedded_doc)
y_fin = np.array(Ytrain)
```

```
x_fin.shape,y_fin.shape
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_fin,y_fin, test_size = 0.33, random_state = 42)
```

**Model Training**

```
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10, batch_size=64)
```
